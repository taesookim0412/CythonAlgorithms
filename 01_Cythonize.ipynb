```python
import os
import contextlib

with open(os.devnull, "w") as f, contextlib.redirect_stdout(f):
    import time
    c__t = time.time()
    [say_hello_to('World') for _ in range(1 * 10**6)]
print(time.time() - c__t)

with open(os.devnull, "w") as f, contextlib.redirect_stdout(f):
    import time
    c__t = time.time()
    [print('Hello, World!') for _ in range(1 * 10**6)]
print(time.time() - c__t)
```
